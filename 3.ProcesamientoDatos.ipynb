```python
import sqlite3
import pandas as pd

df_candidates = pd.read_csv('./src/candidates.csv', sep=';')

conn = sqlite3.connect('./src/candidates.db')

df_candidates.to_sql('candidates', conn, if_exists='replace', index=False)


query ='''
   SELECT 
        AVG("Code Challenge Score") AS Average_Score,
        MAX("Code Challenge Score") AS Max_Score,
        MIN("Code Challenge Score") AS Min_Score,
        AVG("YOE") AS Average_Experience,
        MAX("YOE") AS Max_Experience,
        MIN("YOE") AS Min_Experience
    FROM candidates
'''

results = conn.execute(query).fetchone()

print("Informe de candidatos:")
print(f"Promedio de puntuación del desafío de código: {results[0]}")
print(f"Puntuación máxima del desafío de código: {results[1]}")
print(f"Puntuación mínima del desafío de código: {results[2]}")
print(f"Promedio de años de experiencia: {results[3]}")
print(f"Años de experiencia máxima: {results[4]}")
print(f"Años de experiencia mínima: {results[5]}")

conn.close()
```

## el codigo para envio de correo

```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.base import MIMEBase
from email import encoders

# Datos del remitente
remitente = "johnprivas1@gmail.com"
password = open('clave.txt').read().strip()

# Datos del destinatario
destinatario = "jperaltarivas1@gmail.com"

# Crear el mensaje
mensaje = MIMEMultipart()
mensaje["From"] = remitente
mensaje["To"] = destinatario
mensaje["Subject"] = "Envio de base datos prueba"

# Adjuntar archivo
archivo_adjunto = "./src/candidates.db"  # Ruta y nombre del archivo a adjuntar

# Leer el archivo en modo binario
with open(archivo_adjunto, "rb") as archivo:
    # Crear el objeto MIME base
    adjunto = MIMEBase("application", "octet-stream")
    # Cargar el contenido del archivo
    adjunto.set_payload(archivo.read())

# Codificar el archivo adjunto en base64
encoders.encode_base64(adjunto)

# Establecer las cabeceras del archivo adjunto
adjunto.add_header("Content-Disposition", f"attachment; filename= {archivo_adjunto}")

# Adjuntar el archivo al mensaje
mensaje.attach(adjunto)

# Configurar el servidor SMTP de Gmail
smtp_server = 'smtp.gmail.com'
smtp_port = 587
with smtplib.SMTP(smtp_server, smtp_port) as server:
    
    server.starttls()
    server.login(remitente, password)

    # Enviar el correo electrónico
    server.send_message(mensaje)

```
